```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seaborn library comes with few default data sets in that we are using tips
```

```python
#load first tips dataset

tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.stripplot(y= 'tip', data = tips,jitter=True)
plt.ylabel('tips in $')
plt.show()
```

```python
sns.stripplot(y ='tip' , data = tips , jitter= False)
```

```python
sns.stripplot(x= 'day', y= 'tip' , data=tips)
plt.ylabel('tips in $')
plt.xlabel('day wise ')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
sns.swarmplot(x='day',y='tip', data=tips, hue='sex')
```

```python
sns.swarmplot(x='day',y='tip', data=tips, hue='sex',orient='h')
```

```python
sns.boxplot(x='day',y='tip', data=tips, hue='sex')
```

```python
sns.violinplot(x='day',y='tip', data=tips, hue='sex')
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips  ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips odn ($)')
```

```python
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='black')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips , hue='sex')
```
